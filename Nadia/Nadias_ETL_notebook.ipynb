```python
import pandas as pd
from sqlalchemy import create_engine
import numpy as np
```

# Oscars Dataframe Work

```python
oscars_df = pd.read_csv("../Data/BigML_Dataset_5f50a62c2fb31c516d000176.csv")
oscars_df
```

```python
for col in oscars_df.columns: 
    print(col)
```

```python
copied_oscars_df = oscars_df.drop(oscars_df.filter(regex='nominated').columns, axis=1)
copied_oscars_df
```

```python
dropped_oscars=copied_oscars_df.drop(['movie_id', 'release_date.day-of-month', 'release_date.day-of-week', 'release_date.year'], axis=1)
dropped_oscars
```

```python
dropped_oscars["gross"].dtype
```

```python
no_nans_oscars=dropped_oscars.replace(np.nan, 0)
```

```python
cols = ['gross', 'release_date.month']
no_nans_oscars[cols] = no_nans_oscars[cols].applymap(np.int64)
no_nans_oscars.sample(5)
```

```python
how_many_genres = []
cartoon= []

for index, row in no_nans_oscars.iterrows():
    if 'Animation' in row['genre']:
        cartoon.append(True)
    else:
        cartoon.append(False)
        
    number_of_genres = len(row['genre'].split("|"))
    how_many_genres.append(number_of_genres)

no_nans_oscars['cartoon'] = cartoon
no_nans_oscars['number_of_genres'] = how_many_genres
no_nans_oscars.sample(5)
```

```python
renamed_oscars= no_nans_oscars.rename(columns={'movie':'movie_title'})
```

```python
oscars_df_final=renamed_oscars[["year", "movie_title", "certificate", "duration", "genre", "rate", "metascore", "votes", "gross", "user_reviews", "critic_reviews", "cartoon", "number_of_genres"]].copy(
    ).reset_index(drop=True)
renamed_df
oscars_df_final.head()
```

```python
oscars_df_final.to_csv("../Output/oscars_df_final.csv")
```

```python
nominations_df=renamed_oscars[["year", "movie_title", "awards_wins", "awards_nominations", "Oscar_Best_Picture_won","Oscar_Best_Director_won","Oscar_Best_Actor_won","Oscar_Best_Actress_won","Oscar_Best_Supporting_Actor_won","Oscar_Best_Supporting_Actress_won","Oscar_Best_AdaScreen_won","Oscar_Best_OriScreen_won"]].copy(
    ).reset_index(drop=True)
nominations_df.head()
```

```python
oscar_wins_df=nominations_df.replace(["Yes", "No"], [True, False])
oscar_wins_df.head()
```

```python
oscar_wins_df.to_csv("../Output/oscar_wins_df.csv")
```

# Wiki Dataframe Work

```python
wiki_df = pd.read_csv("../Data/wiki_movie_plots_deduped.csv")
wiki_df
```

```python
filtered_wiki_df = wiki_df.loc[(wiki_df["Origin/Ethnicity"]=="American")]
filtered_wiki_df
```

```python
filtered_wiki_df.columns = filtered_wiki_df.columns.str.replace(' ', '_').str.lower()
renamed_df=filtered_wiki_df.rename(columns={'title':'movie_title', 'origin/ethnicity':'country'}
                                  ).replace({'country': 'American'}, {'country': 'USA'})
renamed_df
```

```python
renamed_wiki_df = renamed_df.loc[(renamed_df["genre"]!="unknown")]
```

```python
duplicates = renamed_wiki_df[renamed_wiki_df.duplicated(['movie_title'])] 
duplicates
```

```python
unique_wiki_df=renamed_wiki_df.drop_duplicates(['movie_title'] , keep ='last') 
unique_wiki_df.shape
```

```python
wiki_df_final=unique_wiki_df[["release_year", "movie_title", "country", "genre", "wiki_page"]].copy(
    ).reset_index(drop=True).dropna(how='all')
wiki_df_final.sample(5)
```

```python
wiki_df_final['release_year'] = wiki_df_final['release_year'].apply(pd.to_numeric)
```

```python
wiki_df_final = wiki_df_final[(wiki_df_final.release_year > 1999)]
```

```python
wiki_df_final
```

```python
wiki_df_final.to_csv("../Output/wiki_clean.csv")
```
