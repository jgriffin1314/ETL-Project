```python
##Imports

import pandas as pd
from sqlalchemy import create_engine
```

```python
##Store CSV into DataFrame

csv_file = "../Data/IMDb movies.csv"
IMBD_df = pd.read_csv(csv_file)
IMBD_df.sample(5)
```

```python
##Read Columns

for col in IMBD_df.columns: 
    print(col)
```

```python
##Cast year in to integer so to sort and drop
IMBD_df[['year']] = IMBD_df[['year']].apply(pd.to_numeric) 
print(IMBD_df.dtypes) 

 
```

```python
##Cast year in to integer so to sort and drop

IMBD_df = IMBD_df[(IMBD_df.country == "USA") & (IMBD_df.year > 1999) & (IMBD_df.year < 2018)] 
IMBD_df.sample(5)
```

```python
## Choose Columns and drop rows where reviews_from_critics is NaN
IMBD_df = IMBD_df.rename(columns={'title': 'movie_title'})
IMBD_df = IMBD_df[['movie_title', 'year', 'genre', 'duration', 'director', 'description', 'avg_vote', 'reviews_from_users', 'reviews_from_critics']].copy()

IMBD_df = IMBD_df[IMBD_df['reviews_from_critics'].notna()]

IMBD_df = IMBD_df.reset_index(drop=True)
IMBD_df.to_csv("../Output/IMBD_clean.csv")

IMBD_df.sample(10)
```

```python
#Connect to Local Database

rds_connection_string = "postgres:postgres@localhost:5432/ETL_project_movie_db"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
#Check for tables
engine.table_names()
```

```python
### Use pandas to load csv converted DataFrame into database
IMBD_df.to_sql(name='imbd_data', con=engine, if_exists='append', index=False)
```

```python
##Confirm data has been added by querying the netflix_data table
pd.read_sql_query('select * from imbd_data', con=engine).head()
```
